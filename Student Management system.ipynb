```python
class Student: 
    def __init__(self, name, stu_id, age, gender, loc): 
        self.__name = name 
        self.__stu_id = stu_id
        self.__age  = age 
        self.__gender = gender
        self.__loc = loc
    def get_id(self):
        return self.__stu_id

    def __str__(self):
        return "Name : "+ self.__name+ "\nstu_id : "+ str(self.__stu_id) + "\nage : "+ str(self.__age) + "\ngender : "+ self.__gender + "\nloc : "+ self.__loc
        print("\n")	 
class Student_management:
    def __init__(self):
        self.__li=[]
    def add(self): 
        Name = input("Enter Name:")
        stu_id = int(input("Enter stu_id:")) 
        age = int(input("Enter age:")) 
        gender = input("Enter gender:") 
        loc = input("Enter loc:")
        ob = Student(Name, stu_id, age, gender, loc ) 
        self.__li.append(ob)

    def display(self): 
        for ob in self.__li:
            print(ob)	 
    def search(self, id): 
        for i in self.__li: 
            if(i.get_id() == id): 
                return i
    def delete(self, si): 
        i = self.search(si) 
        self.__li.remove(i)

ch=1
obj = Student_management()
while(ch!=5):
     

    print("\nOperations: ") 
    print("\n1.Add Student details\n2.Display Student Details\n" 
    "3.Search Details of a Student\n4.Delete Details of a Student" 
     "\n5.Exit") 

    ch = int(input("Enter choice:")) 
    if(ch == 1): 
        obj.add()
    if(ch == 2): 
        obj.display()
    if(ch == 3): 
        id=int(input("Enter id: "))
        x= obj.search(id)
        print(x)
    if(ch==4):
        id=int(input("Enter id: "))
        x= obj.delete(id)
print("Thank you")
```
